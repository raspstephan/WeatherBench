```python
from src.utils import *
import seaborn as sns
sns.set_style('whitegrid')
```

```python
def plot_generalization_error(path, exp_ids):
    exp_ids = [str(exp_id) for exp_id in exp_ids]
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
    colors = sns.palettes.color_palette(n_colors=len(exp_ids))
    names = []
    errors = []
    rel_errors = []
    for exp_id, c, in zip(exp_ids, colors):
        fn = glob(f'{path}{exp_id}*.pkl')[0]
        name = fn.split('/')[-1].split('_history.pkl')[0]
        h = read_pickle(fn)
        train = h['loss'][-1]
        valid = h['val_loss'][-1]
        errors.append(valid-train)
        rel_errors.append((valid-train) / (valid+train) / 2)
        names.append(name)
        
    ax1.bar(names, errors)
    ax2.bar(names, rel_errors)
```

```python
plot_generalization_error('/home/rasp/data/myWeatherBench/predictions/saved_models/', 
            ['71', 86, '81', 85, 84, 83]
        )
```
